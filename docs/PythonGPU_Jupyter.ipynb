The following is how to use one of the ML scicomp machines that
has 4 Titan RTX GPU cards installed. A couple of notes:

1. There are a couple of ways to get the software packages 
setup that are required for using the GPUs. One using a python
virtual environment and the other using a conda environment.

2. Most recently, I have been able to run from a python virtual
environment I created. Check the "Setting up: Python VENV"
section for that.

3. The older way was done using conda. Will Phelps sent me
the magic commands. "See Setting up: Conda" below.

4. It seems at least part of the environment is maintained
when the srun command is called. At least I was still in the
same working directory and python3 was pointing to the venv.
However, the virtual environment was not completely set up
so I had to source the activate script again

5. I was able to pip install from the sciml190X node whereas
it failed to do it from the regular ifarm node due to missing
SSL modules.

Logging in to sciml190X
-------------------------

NOTE: If you have not done the one time installation of the
software into either a Python venv or a Conda virtual
environment, jump down to one of hte sections below for that.

To login to one of the sciml190X nodes do the following:

salloc --gres gpu:TitanRTX:2 --partition gpu --nodes 1<br>
srun --pty bash<br>
export LD_LIBRARY_PATH=/apps/python/3.7.3/lib:/group/12gev_phys/ml/downloads/cudnn-10.0-linux-x64-v7.4.2.24/lib64:${LD_LIBRARY_PATH}


If you are using VENV
-------------------------
source ~/work2/builds/Python_VENV/venv_2020.06.04-gpu/bin/activate

If you are using Conda
-------------------------
source /etc/profile.d/modules.sh<br>
module use /apps/modulefiles<br>
module load anaconda2/4.5.12<br>
conda activate tf-gpu<br>

Other tips on using the sciml nodes
-----------------------------------
If you log out, of both the srun and salloc command, then 
the job should complete and the resource should be released.
You can check this by just running the sacct command to see 
a list of your jobs and if there are any running:

sacct<br>

If there is a running job that you want to kill so the resource is 
released, cancel the jobid via:

scancel *jobid*<br>


To see available devices:

nvidia-smi<br>


To see which devices were assigned to you so this:

printenv CUDA_VISIBLE_DEVICES<br>


Setting up: Conda
-------------------------
This part only needs to be done once to get the software.

The software must be set up using a computer other than sciml190X
since it needs a level of outside network access not available
there. Also, the size of the installation is large enough that
it won't fit easily in you home directory. Conda likes to install
things in ~/.conda so that must be a link to some larger disk.

ssh ifarm1901<br>

\# make a directory on a large disk somewhere and make the ~/.conda directory be a link to it. For example:<br>
mkdir -p /work/halld2/home/davidl/builds/CONDA_ENV<br>
ln -s /work/halld2/home/davidl/builds/CONDA_ENV ~/.conda<br>

bash<br>
